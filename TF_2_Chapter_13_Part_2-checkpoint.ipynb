```python
#created by Dr. Nahid 26/01/2020
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x_train=np.arange(10).reshape((10,1))
y_train=np.array([  1.0, 1.3, 3.1, 2.0, 5.0, 
                    6.3, 6.6, 7.4, 8.0, 9.0])
#print(x_train,y_train)
plt.plot(x_train,y_train,'o',markersize=10)
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.show()
```

```python
#normalize the input data
x_train_norm=(x_train-np.mean(x_train))/np.std(x_train)
#convert the data to tensor domain
x=tf.cast(x_train_norm,tf.float32)
#convert the data to tensor domain
y=tf.cast(y_train,tf.float32)
#data concat in tensor
ds_train_origin=tf.data.Dataset.from_tensor_slices((x,y))
```

```python
#defination of Model

class mymodel(tf.keras.Model):
    def __init__(self):
        super(mymodel, self).__init__()
        self.w = tf.Variable(0.0, name='weight')
        self.b = tf.Variable(0.0, name='bias')
    
    def call(self, x):
        a = self.w*x + self.b
        return a
model = mymodel()
model.build(input_shape=(None, 1))
model.summary()
```

```python
#defination of losses
def loss_fn(y_true,y_predicted):
    return tf.reduce_mean(tf.square(y_true-y_predicted))

yt = tf.convert_to_tensor([1.0])
yp = tf.convert_to_tensor([1.5])
#loss_fn(yt, yp)
```

```python
def train(model, inputs, outputs, learning_rate):
    with tf.GradientTape() as tape:
        current_loss = loss_fn(model(inputs), outputs)
    dW, db = tape.gradient(current_loss, [model.w, model.b])
    model.w.assign_sub(learning_rate * dW)
    model.b.assign_sub(learning_rate * db)
    #model.w=tf.sub(model.w-learning_rate * dW)
    #model.b=tf.sub(model.b-learning_rate * db)
```

```python
#Simple programming for line prediction
tf.random.set_seed(1)
num_epochs = 200
log_steps = 100
learning_rate = 0.001
batch_size = 1
step_per_epoch=np.int(np.ceil(len(y_train))/batch_size)

#shuffling the data and their label
ds_train=ds_train_origin.shuffle(len(y_train))
ds_train=ds_train.repeat(count=None)
ds_train=ds_train.batch(1)
print(ds_train)

#innitial weight and bias defination
Ws, bs= [], [] 

for i, batch in enumerate (ds_train):
    #print(i, ds_train)
    if i >= step_per_epoch * num_epochs:
        break
    #print (i, batch)
    Ws.append(model.w.numpy())
    bs.append(model.w.numpy())
    bx, by = batch
    loss_val = loss_fn(model(bx), by)
    
    train(model, bx, by, learning_rate=learning_rate)
    if i%log_steps==0:
        print('Epoch {:4d} Step {:2d} Loss {:6.4f}'.format(
              int(i/step_per_epoch), i, loss_val))
```

```python
print('Final Parameters', model.w.numpy(), model.b.numpy())
x_test=np.linspace(0,9,num=100).reshape(-1,1)
#print(x_test)
x_test_final=(x_test-np.mean(x_train))/np.std(x_train)
y_pred=model(tf.cast(x_test_final, dtype=tf.float32))

fig = plt.figure(figsize=(13, 5))
ax = fig.add_subplot(1, 2, 1)
plt.plot(x, y, 'o', markersize=10)
plt.plot(x_test_final, y_pred, '--', lw=3)
plt.legend(['Training examples', 'Linear Reg.'], fontsize=15)
ax.set_xlabel('x', size=15)
ax.set_ylabel('y', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)


ax = fig.add_subplot(1, 2, 2)
plt.plot(Ws, lw=3)
plt.plot(bs, lw=3)
plt.legend(['Weight w', 'Bias unit b'], fontsize=15)
ax.set_xlabel('Iteration', size=15)
ax.set_ylabel('Value', size=15)
ax.tick_params(axis='both', which='major', labelsize=15)
#plt.savefig('ch13-linreg-1.pdf')

plt.show()
```

```python
#model training vial .compile and .fit methods
tf.random.set_seed(1)
model = mymodel()
#model.build((None, 1))

model.compile(optimizer='sgd', 
              loss=loss_fn,
              metrics=['mae', 'mse'])
```

```python
# model fiting
model.fit(x_train_norm, y_train, 
          epochs=num_epochs, batch_size=batch_size,
          verbose=1)
```
