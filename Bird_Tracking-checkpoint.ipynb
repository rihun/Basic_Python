```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
birddata = pd.read_csv('D:\Anaconda_practice_file\Bird_Data.csv') 

#birddata = pd.read_csv('bird_tracking.csv') 

bird_names = pd.unique(birddata.bird_name)  
```

```python
bird_names
```

```python
ix = birddata.bird_name == "Eric" 
x,y = birddata.longitude[ix], birddata.latitude[ix] 
plt.figure(figsize = (7,7)) 
plt.plot(x,y,"b.") 
```

```python
plt.figure(figsize = (7,7))
for bird_names1 in bird_names:
    print(bird_names1)
    ix = birddata.bird_name == bird_names1 
    x= birddata.longitude[ix]
    y= birddata.latitude[ix] 
    plt.figure(figsize = (7,7)) 
    plt.plot(x,y,".", label=bird_names1)
plt.xlabel("Longitude") 
plt.ylabel("Latitude") 
plt.legend(loc="lower right") 
plt.show() 
    
```

```python
def bird_tracing(birddata,bird_names2):
    ix = birddata.bird_name == bird_names2 
    x= birddata.longitude[ix]
    y= birddata.latitude[ix] 
    plt.figure(figsize = (7,7)) 
    plt.plot(x,y,".", label=bird_names2)
    plt.xlabel("Longitude") 
    plt.ylabel("Latitude") 
    plt.legend(loc="lower right") 
    plt.show() 
    
    
```

```python
#bird_tracing(birddata,'Eric')

for bird_names1 in bird_names:
    bird_tracing(birddata,bird_names1)
```
